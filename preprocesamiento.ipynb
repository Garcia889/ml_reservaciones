```python
import pandas as pd
import numpy as np
```

```python

#Guarda data/hoteles-entrena.csv en dataframe
df = pd.read_csv('./data/hoteles-entrena.csv')
df.head()
```

```python
#Sustituye children none por 0 y children por 1
df['children'] = df['children'].replace('none', 0)
df['children'] = df['children'].replace('children', 1)

df.head()
```

```python


#Sustituye children none por 0 y children por 1
df['required_car_parking_spaces'] = df['required_car_parking_spaces'].replace('none', 0)
df['required_car_parking_spaces'] = df['required_car_parking_spaces'].replace('parking', 1)

df.head()
```

## Pruebas

```python
df_prueba = df[['required_car_parking_spaces', 'children', 'hotel', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'country', 'booking_changes', 'customer_type']]
df_prueba.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Ejemplo de codificación de variables categóricas
label_encoder = LabelEncoder()
df_prueba['hotel'] = label_encoder.fit_transform(df_prueba['hotel'])

df_prueba.head()
```

```python
# Aplicar OneHotEncoding a la variable 'country'
onehot_encoder = OneHotEncoder()

# Ajustar y transformar las variables categóricas
onehot_encoded = onehot_encoder.fit_transform(df_prueba[['country']]).toarray()

# Obtener los nombres de las nuevas columnas con get_feature_names_out()
column_names = onehot_encoder.get_feature_names_out(['country'])

# Agrergar las nuevas columnas al dataframe original
df_prueba[column_names] = onehot_encoded

df_prueba.head()
```

```python
#Eliminar la columna 'country' original
df_prueba = df_prueba.drop(columns=['country'])

label_encoder = LabelEncoder()
df_prueba['customer_type'] = label_encoder.fit_transform(df_prueba['customer_type'])

df_prueba.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


# Dividir los datos en conjunto de entrenamiento y prueba
X = df_prueba.drop('children', axis=1)  # Variables predictoras
y = df['children']               # Variable objetivo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# Crear el modelo de clasificación
clf = RandomForestClassifier()

# Entrenar el modelo
clf.fit(X_train, y_train)

# Hacer predicciones
y_pred = clf.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print(f'Matriz de Confusión:\n {cm}')

# Reporte de clasificación
report = classification_report(y_test, y_pred)
print(f'Reporte de Clasificación:\n {report}')
```
